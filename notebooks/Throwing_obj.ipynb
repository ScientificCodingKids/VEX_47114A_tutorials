# How to hit a target at given distance?

We can control the velocity and shooting angle to achieve this.


# Fixed angle

Distance increases as velocity increases.


```python
import numpy as np
import pandas as pd
import plotly

pd.options.plotting.backend = "plotly"
```

```python
x = np.arange(0, 50, 0.1)

def height(x: float, v: float, a: float) -> float:
    g = 9.8 # gravity, Newton constant -- the value is made up here!
    t = x / (v * np.cos(a))
    
    h = v * np.sin(a) * t - 0.5 * g * t * t
    
    return h
```

```python
df = pd.DataFrame({"x": x})

df["v=18"] = df["x"].apply(lambda x: height(x, 18, 3.14 / 3))
df["v=20"] = df["x"].apply(lambda x: height(x, 20, 3.14 / 3))
df["v=22"] = df["x"].apply(lambda x: height(x, 22, 3.14 / 3))

df = df.set_index("x")

df.plot(title="Same shooting angle, different speed")


```

# Fixed Speed, varying angles

```python
df2 = pd.DataFrame({"x": x})

df2["60 deg"] = df2["x"].apply(lambda x: height(x, 20, 3.14 / 3))
df2["45 deg"] = df2["x"].apply(lambda x: height(x, 20, 3.14 / 4))
df2["30 deg"] = df2["x"].apply(lambda x: height(x, 20, 3.14 / 6))

df2 = df2.set_index("x")

df2.plot(title="Same speed, varying shooting angles")
```
